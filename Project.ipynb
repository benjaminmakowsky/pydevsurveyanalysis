```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#read data
df17 = pd.read_csv('./pythondevsurvey2017_raw_data.csv')
df18 = pd.read_csv('./python_psf_external_18.csv')
```

```python
df17.drop(labels=['Installer from python.org:What do you typically use to upgrade your Python version?'], axis=1, inplace=True)
```

```python
df17.head(3)
```

```python
df17.rename(columns = {'Is Python the main language you use for your current projects?':'Is Primary'}, inplace=True)
df17.rename(columns=lambda x: x.replace(':What other language(s) do you use?', ''), inplace=True)
df17.rename(columns=lambda x: x.replace(':Which of the following best describes your job role(s)?', ':Job'), inplace=True)
df17.rename(columns=lambda x: x.replace(':\xa0What do you use Python for?', ''), inplace=True)
df18.rename(columns = {'Is Python the main language you use for your current projects?':'Is Primary'}, inplace=True)
df18.rename(columns=lambda x: x.replace(':What other language(s) do you use?', ''), inplace=True)
df18.rename(columns=lambda x: x.replace(':Which of the following best describes your job role(s)?', ':Job'), inplace=True)
df18.rename(columns=lambda x: x.replace(':\xa0What do you use Python for?', ''), inplace=True)
```

```python
list(df17.columns)
```

```python
df17_languages = df17.iloc[:,1:25]
df17_jobs = df17.iloc[:,25:41]
```

```python
#fig1 = sns.barplot(x = df17_languages.count(), y = df17_languages.columns)
lang_counts = df17_languages.count()
lang_counts.sort_values(ascending = True, inplace = True)
plt.figure(figsize=(20,10))
plt.xlabel('Counts')
plt.ylabel('Language')
fig1 = sns.barplot(x = lang_counts, y = lang_counts.index)
```

>As you can see the most common language to know in combination with python is JavaScript. This suggests that UCSD should offer a JavaScript class in their data science program so that the students can be prepared for their careers.

```python
job_counts = df17_jobs.count()
job_counts.sort_values(ascending=True, inplace=True)
plt.figure(figsize=(20,10))
fig2 = sns.barplot(x = job_counts, y = job_counts.index)
```

>However, this data breaks down the field of data science into 3 categories (Machine Learning, Data Analaysis, and Data Gathering). If you combine those 3 fields and remove them from the chart you get a much better representation of how large a role python plays in data science.

```python
dataSum = job_counts.at['Machine learning']+job_counts.at['Data analysis']+job_counts.at['Programming of web parsers / scrapers / crawlers']
job_counts_revised = job_counts.drop(['Machine learning', 'Data analysis', 'Programming of web parsers / scrapers / crawlers'])
job_counts_revised['Data Science'] = dataSum
plt.figure(figsize=(20,10))
fig3 = sns.barplot(x = job_counts_revised, y = job_counts_revised.index)
```
